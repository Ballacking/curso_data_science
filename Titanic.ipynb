```python
# Exportar um Arquivo tipo csv, que seria idêntica ao resultado proposto. Contagem, Name, Age, Sex, Pclass, Survived
# Filtro somente mulheres / Segundo Filtro Pclass, somente primeira classe / Terceiro filtro, ter sobrevivido
# Nomes em Ordem Alfabética
```

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
# A sequência das colunas serão Name, Age,Sex,Pclass e Survived

filter1 = df [['Name','Age','Sex','Pclass','Survived']]
filter1
```

```python
# No caso foi filtrado da tabela que foi executada acima, somente aparecer o sexo feminino, a Pclass como 1 e caso tenha sobrevivido, seria como 1

filter1 = filter1.loc[(filter1['Sex'] == 'female') & (filter1['Pclass'] == 1) & (filter1['Survived'] == 1)]
filter1
```

```python
# Colocando a ordem alfabetica nos nomes

filter1 = filter1.sort_values (['Name'], ascending = True)
filter1
```

```python
# Convertendo o arquivo 

filter1.to_csv('Titanic_Solution.csv', index = False)
```
