```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.loc[df['Survived'] == 1, 'Survived'] = 'Yes' 
df
```

```python
df.loc[df['Survived'] == 0 , 'Survived'] = 'No' 
df
```

```python
filter1 = df [['Name','Age','Sex','Pclass','Survived']]
filter1
```

```python
filter1 = filter1.loc[(filter1['Sex'] == 'female') & (filter1['Pclass'] == 3) & (filter1['Survived'] == 'No')]
filter1
```

```python
filter1.to_csv('Titanic_Teste.csv', index = False)
```
