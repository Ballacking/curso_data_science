```python
# Exportar um Arquivo tipo csv, que seria idêntica ao resultado proposto. Contagem, Name, Age, Sex, Pclass, Survived
# Filtro somente mulheres / Segundo Filtro Pclass, somente primeira classe / Terceiro filtro, ter sobrevivido
# Nomes em Ordem Alfabética
```

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
filter1 = df [['Name','Age','Sex','Pclass','Survived', 'Cabin']]
filter1
```

```python
filter1 = filter1.loc[(filter1['Sex'] == 'male') & (filter1['Pclass'] == 3 ) & (filter1['Age'] <= 15)]
filter1
```

```python
filter1 = filter1.sort_values (['Age'], ascending = True)
filter1
```

```python
filter1.to_csv('Titanic_Solution_2.csv', index = False)
```
